# Гэрийн даалгавар:
### Жингийн илүүдэлтэй эсэхийг олдог программ бич. IF функцийг ашиглана. (Өндөр / жин)


```python
undur = float(input('Биеийн өндөр оруулна уу (cm): '))
jin = float(input('Биеийн жин оруулна уу (кг): '))

if undur <50 or undur >250:
  print(" Алдаа: Өндөр 50-250 см хооронд байх ёстой!")
elif jin < 10 or jin > 300:
    print(" Алдаа: Жин 10-300 кг хооронд байх ёстой!")

else:
  undur = undur / 100
  x = jin / (undur ** 2)
  print(f'Таны илүүдэл жин: {x:.1f}')
  if x <= 18.5:
    print("Байдал: Жин дутуу!")
  elif x < 25.0:
      print("Байдал: Хэвийн жин")
  elif x < 30.0:
      print("Байдал: Жин илүүдэлтэй")
  else:
      print("Байдал: Таргалалт")
```
